```python
import os

# mapping of old folder names to new ones
rename_map = {
    "deu": "german",
    "pol": "polish",
    "eng": "english",
    "fra": "french",
    "lav": "latvian",
    "ita": "italian",
    "lit": "lithuanian",
    "mlt": "maltese",
    "dsb": "upper_sorbian"
    }
  

# root directory where your folders live
dir_list = ["derivation/task-2/3_shot_examples", 
        "derivation/task-2/5_shot_examples",
        "derivation/task-2/10_shot_examples"
    ]
for root_dir in dir_list:
    print("Renaming dirictory path ", root_dir)
    for dirpath, dirnames, filenames in os.walk(root_dir, topdown=True):
        for old, new in rename_map.items():
            if old in dirnames:
                old_path = os.path.join(dirpath, old)
                new_path = os.path.join(dirpath, new)
                print(old_path)
                print(new_path)
                print(f"Renaming {old_path} -> {new_path}")
                os.rename(old_path, new_path)
                # update dirnames so os.walk doesnâ€™t get confused
                dirnames[dirnames.index(old)] = new
```

```python
import os
import glob
import shutil

base_dir = "syntax/5_shot_examples"

for lang in os.listdir(base_dir):
    lang_dir = os.path.join(base_dir, lang)
    # print(lang_dir)
    if not os.path.isdir(lang_dir):
        continue

    # match test* and dev* files directly inside each lang folder
    patterns = ["few_shot_test*", "few_shot_dev*"]
    for pat in patterns:
        for file in glob.glob(os.path.join(lang_dir, pat)):
            print(file)
            try:
                os.remove(file)
                print(f"Removed file: {file}")
            except IsADirectoryError:
                shutil.rmtree(file)
                print(f"Removed directory: {file}")
```
